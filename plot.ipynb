```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
SMALL_SIZE = 30
MEDIUM_SIZE = 30
BIGGER_SIZE = 40

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
ddim_deterministic = [488.20760778462875, 439.32746628785287, 441.0980578459961, 440.50303956210655, 466.6483630332974]
ddim_deterministic_speedup = [460.6516529450798, 472.1551978889844, 457.1643780284994, 455.914908917716, 464.5819311114063]
ddim_stochastic = [553.9195259147989, 541.397109620301, 535.9680133193701, 528.0248535431376, 540.2200945350758]
ddpm_stochastic = [646.9972807310756, 612.4844075724457, 655.1502356422892, 645.4002962310738, 665.0936646634392]

means_ddim_deterministic = np.mean(ddim_deterministic)
means_ddim_deterministic_speedup = np.mean(ddim_deterministic_speedup)
means_ddim_stochastic = np.mean(ddim_stochastic)
means_ddpm_stochastric = np.mean(ddpm_stochastic)

stds_ddim_deterministic = np.std(ddim_deterministic)
stds_ddim_deterministic_speedup = np.std(ddim_deterministic_speedup)
stds_ddim_stochastic = np.std(ddim_stochastic)
stds_ddpm_stochastric = np.std(ddpm_stochastic)


model = ["ddim\ndeterm", "ddim\ndeterm\nspeedup", "ddim\nstochastic", "ddpm"]
means = [means_ddim_deterministic,means_ddim_deterministic_speedup, means_ddim_stochastic, means_ddpm_stochastric]
stds = [stds_ddim_deterministic, stds_ddim_deterministic_speedup, stds_ddim_stochastic, stds_ddpm_stochastric]
plt.figure(figsize=(10,6))
plt.bar(model, means)
plt.errorbar(model, means, yerr=stds, fmt="o", color="r")
plt.ylabel('MSE')
plt.ylim([300,700])

stds
```
